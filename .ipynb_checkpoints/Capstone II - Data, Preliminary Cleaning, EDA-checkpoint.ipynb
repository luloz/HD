# Demographics Data Dictionary
     'SEQN' - Respondent sequence number. 
     'RIAGENDR' - Gender of participant
     'RIDAGEYR' - Age in Year, top coded 80
     'RIDAGEMN' - Age in months at time of screening, top coded 959 SO MANY NULLS
     'RIDRETH1' - Recode of reported race and Hispanic origin information
     'RIDRETH3' - Recode of reported race and Hispanic origin information, with Non-Hispanic Asian Category
     'RIDEXMON' - Six month time period when the examination was performed - two categories: November 1 through April 30, May 1 through October 31.
     'RIDEXAGM' - Age in months at time of examination
     'DMQMILIZ' - Military Service 1.0 = yes, 2.0 = no, 7.0 error? 
     'DMQADFC' - Served in foreign country during military service
     'DMDBORN4' - What country were you born, seems like the vast majority are 1 which probably means US
     'DMDCITZN' - Citizenship? 
     'DMDYRSUS' - Length of time the participant has been in the US.
     'DMDEDUC3' - 
     'DMDEDUC2',
     'DMDMARTL',
     'RIDEXPRG',
     'SIALANG' - Language used for sampling
     'SIAPROXY' - language proxy
     'SIAINTRP' - interpretor used
     'FIALANG' - family interview language
     'FIAPROXY' - proxy used for family interview
     'FIAINTRP' - interpretor used for family
     'MIALANG',
     'MIAPROXY',
     'MIAINTRP',
     'AIALANGA',
     'DMDHHSIZ' - Total number of people in household 
     'DMDFMSIZ' - Total number of people in family
     'DMDHHSZA'- num children >=5 in household
     'DMDHHSZB' - num children 6-17 in household
     'DMDHHSZE',
